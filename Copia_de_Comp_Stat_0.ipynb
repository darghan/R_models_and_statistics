$\mathcal{F}[f(t)](\omega) = \int_{-\infty}^{\infty} f(t) e^{-i\omega t} \, dt$

$$
\begin{cases}
2x + 3y = 8 \\
x - y = 1
\end{cases}
$$

```python
import numpy as np

# Define la matriz de coeficientes y el vector de términos independientes
A = np.array([[2, 3], [1, -1]])
b = np.array([8, 1])

# Resuelve el sistema de ecuaciones
x = np.linalg.solve(A, b)

# Imprime la solución
print("x =", x[0])
print("y =", x[1])
```
